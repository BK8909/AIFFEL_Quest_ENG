# Q1-1 회문 판독기

```python
def is_palindrome(text):

    reverse_text = text[::-1]

    if text == reverse_text:
        return True

    else:
        return False


target1 = "아많다많다많다많아"
target2 = "아정말많다많다많아"

print(is_palindrome(target1))
print(is_palindrome(target2))
```

# Q1-2 회문 분류기

```python
def is_palindrome(text):

    reverse_text = text[::-1]

    return text == reverse_text


def palindrome_classifier(targets):

    result = {
        "회문O": [],
        "회문X": []
    }

    palindrome_count = 0

    for word in targets:

        if is_palindrome(word):

            result["회문O"].append(word)
            palindrome_count += 1

        else:

            result["회문X"].append(word)

    palindrome_ratio = palindrome_count / len(targets)

    return result, palindrome_count, palindrome_ratio


targets = [
    '기러기',
    '까마귀',
    '토마토',
    '스윗소스',
    '스위스',
    '인도인',
    '한국인',
    '별똥별',
    '우영우',
    '우향우'
]

print(palindrome_classifier(targets))
```

# Q2-2 성적 분석기

```python
scores = {
    "유원영": {
        "국어": 94,
        "영어": 91,
        "수학": 89
    },

    "설지민": {
        "국어": 88,
        "영어": 96,
        "수학": 90
    },

    "신윤아": {
        "국어": 86,
        "영어": 88,
        "수학": 77
    },

    "장윤아": {
        "국어": 98,
        "영어": 76,
        "수학": 92
    }
}


def score_analysis(name):

    subjects = ["국어", "영어", "수학"]

    print(f"[{name} 성적 분석]")

    student_total = 0
    class_total = 0

    for subject in subjects:

        student_score = scores[name][subject]

        total = 0

        for student in scores:
            total += scores[student][subject]

        average = total / len(scores)

        diff = student_score - average

        print(
            f"{subject} {student_score}점, "
            f"학급 평균({average:.2f}점) 대비 {diff:+.2f}"
        )

        student_total += student_score
        class_total += average

    student_avg = student_total / len(subjects)

    class_avg = class_total / len(subjects)

    avg_diff = student_avg - class_avg

    print()

    print(
        f"전과목 평균 : {student_avg:.2f}점, "
        f"학급 평균({class_avg:.2f}점) 대비 {avg_diff:+.2f}"
    )


score_analysis("설지민")
```
