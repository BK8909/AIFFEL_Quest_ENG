# 클로저와 데코레이터 퀘스트

## Q1

```python
# 최솟값과 최댓값을 저장하는 클로저 함수 만들기

def find_min_max(numbers):

    # 지역 변수로 최솟값과 최댓값 저장
    min_value = min(numbers)
    max_value = max(numbers)

    # 최솟값 반환 함수
    def find_min():
        return min_value

    # 최댓값 반환 함수
    def find_max():
        return max_value

    # 내부 함수 반환
    return find_min, find_max

# 숫자 리스트 생성
numbers = [10, 5, 8, 12, 3, 7]

# 반환된 함수 저장
find_min, find_max = find_min_max(numbers)

# 결과 출력
print('최솟값:', find_min())
print('최댓값:', find_max())
```

## Q2

```python
# 함수 실행 횟수를 저장하는 데코레이터 만들기

def counter(func):

    # 함수 실행 횟수를 저장할 변수 생성
    count = 0

    # 내부 함수 생성
    # *args : 여러 개의 위치 인자를 받을 수 있음
    # **kwargs : 키워드 인자를 받을 수 있음
    # 매개변수가 있는 함수도 처리 가능하도록 작성
    def wrapper(*args, **kwargs):

        # 바깥 함수의 count 변수 수정
        nonlocal count

        # 함수 실행 횟수 증가
        count += 1

        # 원래 함수 실행
        # 전달받은 인자들을 그대로 func 함수에 전달
        result = func(*args, **kwargs)

        # 현재 실행 중인 함수 이름과 실행 횟수 출력
        print(f'{func.__name__} 실행횟수 : {count}')

        # 원래 함수의 반환값 유지
        return result

    # 내부 함수 반환
    # wrapper 함수가 count 값을 기억하는 클로저 구조
    return wrapper

# @counter 적용
# hello = counter(hello) 와 같은 의미
@counter
def hello():

    # hello 함수 실행 시 출력
    print('Hello Aiffel!')

# 함수 반복 실행
hello()
hello()
hello()
hello()
hello()
```

## 디버깅 내용

- 처음에는 wrapper() 함수에 매개변수를 작성하지 않아 매개변수가 있는 함수에서는 오류가 발생할 수 있었다.
- wrapper(*args, **kwargs) 형태로 수정하여 다양한 함수에서도 사용할 수 있도록 개선하였다.
- f-string 따옴표 오류와 들여쓰기 오류를 수정하면서 파이썬의 문법과 들여쓰기의 중요성을 다시 확인할 수 있었다.

## 회고

데코레이터 구조가 처음에는 굉장히 복잡하게 느껴졌다.
특히 @counter 와 return wrapper 의 흐름이 가장 헷갈렸다.

하지만 hello = counter(hello) 와 같은 구조라는 점을 이해하면서 조금씩 흐름을 파악할 수 있었다.

또한 nonlocal을 사용하는 이유와 wrapper 함수가 count 값을 기억하는 클로저 구조라는 점도 새롭게 배울 수 있었다.
