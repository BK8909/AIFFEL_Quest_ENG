# [QUEST 04] 예외 처리를 활용한 계산기 프로그램 만들기

## 프로젝트 개요
사용자로부터 두 개의 정수와 연산자(`+`, `-`, `*`, `/`, `**`)를 입력받아  
계산 결과를 출력하는 계산기 프로그램을 구현하였습니다.

## 진행 방식
- **Driver** : 천세문
- **Navigator** : 강경수

## 구현 조건
- 정수가 아닌 값을 입력할 경우 `ValueError` 예외 처리
- 올바른 정수를 입력할 때까지 재입력 요청
- 0으로 나누는 경우 `ZeroDivisionError` 처리
- 지원하지 않는 연산자 입력 시 오류 메시지 출력
- `math` 모듈을 활용한 제곱 연산 지원
- 계산 반복 여부 선택 기능 구현


````python
# math 모듈 불러오기
import math


# 정수를 안전하게 입력받는 함수
def input_number(message):

    while True:

        try:
            # 사용자 입력값을 정수형으로 변환
            number = int(input(message))

            return number

        # 정수가 아닌 값을 입력한 경우 예외 처리
        except ValueError:

            print("잘못된 입력입니다. 정수를 입력해주세요.")


# 계산기 프로그램 반복 실행
while True:

    # 첫 번째 정수 입력
    num1 = input_number("첫 번째 정수를 입력하세요 : ")

    # 두 번째 정수 입력
    num2 = input_number("두 번째 정수를 입력하세요 : ")

    # 연산자 입력
    operator = input("연산자를 입력하세요 (+, -, *, /, **) : ")

    try:

        # 덧셈
        if operator == "+":

            result = num1 + num2

        # 뺄셈
        elif operator == "-":

            result = num1 - num2

        # 곱셈
        elif operator == "*":

            result = num1 * num2

        # 나눗셈
        elif operator == "/":

            # 두 번째 숫자가 0인지 확인
            if num2 == 0:

                raise ZeroDivisionError("0으로 나눌 수 없습니다.")

            result = num1 / num2

        # 제곱 연산
        elif operator == "**":

            # math.pow()를 사용한 거듭제곱 계산
            result = math.pow(num1, num2)

        # 지원하지 않는 연산자 입력 시 예외 발생
        else:

            raise ValueError("지원하지 않는 연산자입니다.")

        # 계산 결과 출력
        print(f"계산 결과 : {result}")

    # 0으로 나누기 오류 처리
    except ZeroDivisionError as e:

        print(f"오류 : {e}")

    # 잘못된 입력 또는 연산자 오류 처리
    except ValueError as e:

        print(f"오류 : {e}")

    # 예상하지 못한 오류 처리
    except Exception as e:

        print(f"예상치 못한 오류가 발생했습니다 : {e}")

    # 계속 계산 여부 입력
    again = input("계속 계산하시겠습니까? (y/n) : ")

    # y가 아니면 프로그램 종료
    if again.lower() != "y":

        print("계산기를 종료합니다.")

        break
````

# 오류 수정 및 개선 사항

## 1. SyntaxError 수정
### 발생 오류
```python
SyntaxError: expected 'except' or 'finally' block
```

### 원인
- `try` 블록과 `except` 블록의 들여쓰기가 올바르지 않았음
- 주석 처리 과정에서 코드 구조가 깨짐

### 수정 내용
- `try-except` 구조를 올바르게 재정렬
- 들여쓰기 통일

---

## 2. ZeroDivisionError 처리 개선

### 수정 전
```python
result = num1 / num2
if num2 == 0:
    raise ZeroDivisionError
```

### 수정 후
```python
if num2 == 0:
    raise ZeroDivisionError("0으로 나눌 수 없습니다.")
result = num1 / num2
```

### 개선 사항
- 0인지 먼저 확인 후 나눗셈 수행
- 사용자에게 더 명확한 오류 메시지 제공
