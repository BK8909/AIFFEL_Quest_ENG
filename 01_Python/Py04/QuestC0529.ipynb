## [퀘스트 1번: 은행계좌 만들기]

### 프로젝트 개요
은행 계좌 생성, 입출금, 이자 지급, 입출금 내역 관리 기능을 구현하였다.

```python
import random

# ==================================================
# 퀘스트 1번 : 은행 계좌(Account) 클래스 구현
# ==================================================

class Account:
    # Q2. 생성된 계좌 개수를 저장하는 클래스 변수
    account_count = 0

    # Q1. 계좌 생성자
    def __init__(self, owner, balance):
        self.bank_name = "SC은행"      # 은행명
        self.owner = owner             # 예금주
        self.balance = balance         # 현재 잔고

        # 계좌번호 자동 생성
        self.account_number = self._generate_account_number()

        # Q7. 입금 횟수 저장 (5회마다 이자 지급)
        self._deposit_count = 0

        # Q10. 입금/출금 내역 저장용 리스트
        self._deposit_log = []
        self._withdraw_log = []

        # 계좌 생성 시 계좌 수 증가
        Account.account_count += 1

    # 계좌번호 생성 메서드 (형식 : 000-00-000000)
    def _generate_account_number(self):
        part1 = str(random.randint(100, 999))
        part2 = str(random.randint(10, 99))
        part3 = str(random.randint(100000, 999999))

        return f"{part1}-{part2}-{part3}"

    # Q3. 현재 생성된 계좌 개수 출력
    @classmethod
    def get_account_num(cls):
        print(f"현재 생성된 계좌 수: {cls.account_count}개")

    # Q4. 입금 기능
    def deposit(self, amount):

        # 1원 미만 입금 불가
        if amount < 1:
            print("입금은 최소 1원 이상만 가능합니다.")
            return

        # 잔고 증가
        self.balance += amount

        # 입금 횟수 증가
        self._deposit_count += 1

        # 입금 내역 저장
        self._deposit_log.append(amount)

        print(f"{amount:,}원이 입금되었습니다. 잔고: {self.balance:,}원")

        # Q7. 입금 5회마다 잔고의 1% 이자 지급
        if self._deposit_count % 5 == 0:
            interest = int(self.balance * 0.01)
            self.balance += interest

            print(
                f"🎉 입금 5회 달성! "
                f"이자 {interest:,}원이 지급되었습니다. "
                f"잔고: {self.balance:,}원"
            )

    # Q5. 출금 기능
    def withdraw(self, amount):

        # 출금 금액이 잔고보다 크면 출금 불가
        if amount > self.balance:
            print("잔고가 부족합니다. 출금할 수 없습니다.")
            return

        # 잔고 감소
        self.balance -= amount

        # 출금 내역 저장
        self._withdraw_log.append(amount)

        print(f"{amount:,}원이 출금되었습니다. 잔고: {self.balance:,}원")

    # Q6. 계좌 정보 출력
    def display_info(self):
        print(
            f"은행이름: {self.bank_name}, "
            f"예금주: {self.owner}, "
            f"계좌번호: {self.account_number}, "
            f"잔고: {self.balance:,}원"
        )

    # Q10. 입금 내역 출력
    def deposit_history(self):
        print(f"[{self.owner}] 입금 내역:")

        if not self._deposit_log:
            print("  입금 내역이 없습니다.")

        for i, amount in enumerate(self._deposit_log, 1):
            print(f"  {i}회: +{amount:,}원")

    # Q10. 출금 내역 출력
    def withdraw_history(self):
        print(f"[{self.owner}] 출금 내역:")

        if not self._withdraw_log:
            print("  출금 내역이 없습니다.")

        for i, amount in enumerate(self._withdraw_log, 1):
            print(f"  {i}회: -{amount:,}원")


# ==================================================
# 테스트 코드
# ==================================================

print("=" * 50)
print("퀘스트 1번 : 은행 계좌 만들기")
print("=" * 50)

# Q8. 계좌 객체 4개 생성
acc1 = Account("김철수", 500000)
acc2 = Account("이영희", 1500000)
acc3 = Account("박민준", 2000000)
acc4 = Account("최수연", 800000)

# 계좌 객체를 리스트에 저장
accounts = [acc1, acc2, acc3, acc4]

# Q3. 현재 생성된 계좌 개수 확인
Account.get_account_num()

print()

# Q6. 전체 계좌 정보 출력
print("[계좌 정보]")

for acc in accounts:
    acc.display_info()

print()

# Q4 / Q5. 입출금 기능 테스트
print("[입출금 테스트 - 김철수 계좌]")

acc1.deposit(100000)
acc1.deposit(50000)
acc1.deposit(200000)

acc1.withdraw(30000)

acc1.deposit(10000)
acc1.deposit(20000)   # 5번째 입금 → 이자 지급

print()

# Q10. 입금/출금 내역 확인
acc1.deposit_history()
acc1.withdraw_history()

print()

# Q9. 잔고가 100만원 이상인 고객 정보 출력
print("[잔고 100만원 이상 고객]")

for acc in accounts:
    if acc.balance >= 1000000:
        acc.display_info()
```

### 오류 수정 및 개선 사항
입금 5회마다 이자가 지급되도록 입금 횟수를 저장하는 변수를 추가하였다.

계좌번호 생성 기능은 별도 메소드로 분리하여 가독성을 높였다.

### 회고
클래스 변수와 인스턴스 변수의 차이를 이해할 수 있었고, 메소드를 기능별로 분리하는 방법을 연습할 수 있었다.

## [퀘스트 2번: 간단한 자동사냥 RPG 만들기]

### 프로젝트 개요
상속을 활용하여 Player와 Monster를 구현하고 전투, 경험치, 레벨업 기능을 구현하였다.

```python
import random

# ==================================================
# 퀘스트 2번 : 간단한 자동사냥 RPG 만들기
# ==================================================


# ==================================================
# Q1. Character 클래스
# 모든 캐릭터(플레이어, 몬스터)의 공통 기능 정의
# ==================================================
class Character:

    # 캐릭터 기본 정보 초기화
    def __init__(self, name, level, hp, attack, defense):
        self.name = name          # 이름
        self.level = level        # 레벨
        self.hp = hp              # 체력
        self.attack = attack      # 공격력
        self.defense = defense    # 방어력

    # 현재 생존 여부 확인
    # 체력이 0보다 크면 True 반환
    def is_alive(self):
        return self.hp > 0

    # 공격을 받았을 때 체력 감소 처리
    def take_damage(self, damage):

        # 실제 데미지 계산
        # (공격 데미지 - 방어력)
        actual_damage = damage - self.defense

        # 실제 데미지가 0보다 클 때만 체력 감소
        if actual_damage > 0:
            self.hp -= actual_damage

        # 방어력이 더 높으면 체력 감소 없음

    # 대상(target)을 공격하는 메서드
    def attack_target(self, target):

        # 1 ~ 공격력 사이의 랜덤 데미지 생성
        damage = random.randint(1, self.attack)

        # 대상에게 데미지 적용
        target.take_damage(damage)

        # 공격 결과 출력
        print(
            f"{self.name}이(가) "
            f"{target.name}에게 "
            f"{damage}만큼 공격했다...!"
        )

        print(f"{target.name}의 체력: {target.hp}")


# ==================================================
# Q2. Player 클래스
# Character를 상속받아 플레이어 기능 구현
# ==================================================
class Player(Character):

    # 플레이어 생성
    def __init__(self, name):

        # 기본 능력치 설정
        super().__init__(
            name=name,
            level=1,
            hp=100,
            attack=25,
            defense=5
        )

        # 경험치 속성 추가
        self.exp = 0

    # 경험치 획득
    def gain_experience(self, amount):

        self.exp += amount

        print(
            f"{self.name}이(가) "
            f"경험치 {amount}를 획득했습니다! "
            f"(현재 경험치: {self.exp})"
        )

    # 레벨업 처리
    def level_up(self):

        # 경험치 50당 1레벨 상승
        while self.exp >= 50:

            # 사용한 경험치 차감
            self.exp -= 50

            # 레벨 증가
            self.level += 1

            # 능력치 증가
            self.attack += 10
            self.defense += 5

            print(
                f"🎉 레벨 업! "
                f"{self.name}의 레벨이 "
                f"{self.level}이 되었습니다! "
                f"(공격력: {self.attack}, "
                f"방어력: {self.defense})"
            )


# ==================================================
# Q2. Monster 클래스
# Character를 상속받아 몬스터 생성
# ==================================================
class Monster(Character):

    # 몬스터 생성
    def __init__(self, name, level):

        # 레벨에 비례하여 능력치 랜덤 생성
        hp = random.randint(10, 30) * level
        attack = random.randint(5, 20) * level
        defense = random.randint(1, 5) * level

        super().__init__(
            name=name,
            level=level,
            hp=hp,
            attack=attack,
            defense=defense
        )


# ==================================================
# Q3. 전투 함수
# 플레이어와 몬스터가 체력이 0 이하가 될 때까지 전투
# ==================================================
def battle(player, monster):

    print(f"\n{'=' * 40}")
    print(f"{monster.name} 과의 전투를 시작합니다.")
    print(f"{'=' * 40}")

    # 둘 다 살아있는 동안 반복
    while player.is_alive() and monster.is_alive():

        # 플레이어 선공
        player.attack_target(monster)

        # 몬스터가 죽었으면 전투 종료
        if not monster.is_alive():
            break

        # 몬스터 반격
        monster.attack_target(player)

    # 플레이어 승리
    if player.is_alive():

        # 몬스터 레벨 × 20 만큼 경험치 획득
        exp_gain = monster.level * 20

        player.gain_experience(exp_gain)

        # 레벨업 여부 확인
        player.level_up()

        print("전투 승리!")
        print(f"{monster.name} 을(를) 이겼다!")

    # 플레이어 패배
    else:
        print("전투 패배..")


# ==================================================
# Q4. 게임 실행 함수
# ==================================================
def main():

    # 등장 몬스터 목록
    # key : 몬스터 이름
    # value : 몬스터 레벨
    monster_dict = {
        '슬라임': 1,
        '고블린': 2,
        '오크': 3
    }

    # 플레이어 생성
    player_name = input("플레이어 이름을 입력하세요: ")

    player = Player(player_name)

    print(
        f"\n{player.name} 님, 게임을 시작합니다! "
        f"(체력: {player.hp}, "
        f"공격력: {player.attack}, "
        f"방어력: {player.defense})\n"
    )

    # 몬스터를 순서대로 생성하여 전투 진행
    for monster_name, monster_level in monster_dict.items():

        # 몬스터 생성
        monster = Monster(
            monster_name,
            monster_level
        )

        # 전투 시작
        battle(player, monster)

        # 플레이어 사망 시 게임 종료
        if not player.is_alive():
            print("\n게임오버")
            return

    # 모든 몬스터 처치 성공
    print(
        f"\n🏆 모든 몬스터를 물리쳤습니다! "
        f"최종 레벨: {player.level}"
    )


# ==================================================
# 프로그램 시작 지점
# ==================================================
if __name__ == "__main__":
    main()
```

### 오류 수정 및 개선 사항
super().__init__()를 이용한 부모 생성자 호출 구조를 이해하고 적용하였다.

is_alive() 메소드를 활용하여 전투 종료 조건을 구현하였다.
