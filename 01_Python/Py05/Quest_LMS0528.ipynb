## 1. 사각형 넓이 구하기

직사각형(rectangle), 평행사변형(parallelogram), 사다리꼴(trapezoid)의 넓이를 구하는 클래스입니다.


```python
# 사각형 넓이 구하기 클래스 생성
class Square:

    # 생성자 메소드
    # 객체 생성 시 자동 실행
    def __init__(self):

        # 사용자에게 도형 선택 받기
        self.square = int(input(
            '넓이를 구하고 싶은 사각형의 숫자를 입력하세요.\n'
            '1. 직사각형  2. 평행사변형  3. 사다리꼴\n>>> '
        ))

        # 직사각형 선택
        if self.square == 1:

            print('직사각형 함수는 rect() 입니다.')

            # rect 메소드 실행
            print(self.rect())

        # 평행사변형 선택
        elif self.square == 2:

            print('평행사변형 함수는 par() 입니다.')

            # par 메소드 실행
            print(self.par())

        # 사다리꼴 선택
        elif self.square == 3:

            print('사다리꼴 함수는 trape() 입니다.')

            # trape 메소드 실행
            print(self.trape())

        # 잘못 입력했을 경우
        else:

            print('1, 2, 3 중에서 다시 입력해주세요')


    # 직사각형 넓이 메소드
    def rect(self):

        # 가로, 세로 입력
        width, vertical = map(
            int,
            input(
                '가로, 세로를 입력하세요. 예시 : 가로,세로\n>>> '
            ).split(',')
        )

        # 넓이 계산
        area = width * vertical

        # 결과 문자열 생성
        result = '직사각형의 넓이는 : ' + str(area)

        # 결과 반환
        return result


    # 평행사변형 넓이 메소드
    def par(self):

        # 밑변, 높이 입력
        lower, height = map(
            int,
            input(
                '밑변, 높이를 입력하세요. 예시 : 밑변,높이\n>>> '
            ).split(',')
        )

        # 넓이 계산
        area = lower * height

        # 결과 문자열 생성
        result = '평행사변형의 넓이는 : ' + str(area)

        # 결과 반환
        return result


    # 사다리꼴 넓이 메소드
    def trape(self):

        # 밑변, 윗변, 높이 입력
        lower, upper, height = map(
            int,
            input(
                '밑변, 윗변, 높이를 입력하세요. 예시 : 밑변,윗변,높이\n>>> '
            ).split(',')
        )

        # 넓이 계산
        area = (lower + upper) * height / 2

        # 결과 문자열 생성
        result = '사다리꼴의 넓이는 : ' + str(area)

        # 결과 반환
        return result


# 객체 생성
a = Square()
```


## 2. 키오스크 만들기

메뉴 출력, 주문, 추가 주문, 주문표 출력, 결제 기능이 포함된 키오스크 클래스입니다.


```python
# 키오스크(Kiosk) 클래스 생성
class Kiosk:

    # 생성자 메소드
    # 객체가 생성될 때 자동 실행됨
    def __init__(self):

        # 메뉴 이름 리스트
        self.menu = [
            'Americano',
            'Latte',
            'Mocha',
            'Yuzu_Tea',
            'Green_Tea',
            'Choco_Latte'
        ]

        # 메뉴 가격 리스트
        self.price = [5800, 6800, 7200, 7500, 7300, 7800]

        # 주문한 메뉴를 저장하는 리스트
        self.order_menu = []

        # 주문한 메뉴 가격을 저장하는 리스트
        self.order_price = []

        # 총 주문 금액 저장 변수
        self.price_sum = 0


    # 메뉴 출력 메소드
    def menu_print(self):

        # 메뉴판 상단 출력
        print("=" * 40)

        # 카페 이름 가운데 정렬
        print("Maison de Café".center(40))

        print("=" * 40)

        # 메뉴 개수만큼 반복
        for i in range(len(self.menu)):

            # 메뉴 번호 / 메뉴 이름 / 가격 출력
            print(f"{i+1}. "
                  f"{self.menu[i]:<20}"
                  f"{self.price[i]:>6}원")

        # 메뉴판 하단 출력
        print("=" * 40)


    # 주문 메소드
    def menu_select(self):

        # 첫 주문 입력
        n = int(input("\n음료 번호를 입력하세요 : "))

        # 메뉴 번호가 범위를 벗어나면 반복 입력
        while n < 1 or n > len(self.menu):

            print("없는 메뉴입니다.")

            n = int(input("다시 입력하세요 : "))

        # HOT / ICE 선택 변수
        t = 0

        # 1 또는 2를 입력할 때까지 반복
        while t != 1 and t != 2:

            # HOT / ICE 입력
            t = int(input(
                "HOT 음료는 1을, ICE 음료는 2를 입력하세요 : "
            ))

            # HOT 선택
            if t == 1:
                self.temp = "HOT"

            # ICE 선택
            elif t == 2:
                self.temp = "ICE"

            # 잘못 입력했을 경우
            else:
                print("다시 입력하세요.")

        # 주문 문자열 생성
        order = self.temp + ' ' + self.menu[n - 1]

        # 주문 리스트에 메뉴 추가
        self.order_menu.append(order)

        # 주문 가격 리스트에 가격 추가
        self.order_price.append(self.price[n - 1])

        # 총 금액에 가격 추가
        self.price_sum += self.price[n - 1]

        # 첫 주문 출력
        print()

        print("주문 음료",
              order,
              ':',
              self.price[n - 1],
              '원')

        print("합계 :",
              self.price_sum,
              '원')


        # 추가 주문 반복
        while n != 0:

            print()

            # 추가 주문 또는 결제 입력
            n = int(input(
                "추가 주문은 음료 번호를, 지불은 0을 누르세요 : "
            ))

            # 0 입력 시 반복 종료
            if n == 0:
                break

            # 없는 메뉴 번호 입력 시
            if n < 1 or n > len(self.menu):

                print("없는 메뉴입니다.")
                continue

            # HOT / ICE 선택 변수 초기화
            t = 0

            # 1 또는 2 입력 전까지 반복
            while t != 1 and t != 2:

                # HOT / ICE 입력
                t = int(input(
                    "HOT 음료는 1을, ICE 음료는 2를 입력하세요 : "
                ))

                # HOT 선택
                if t == 1:
                    self.temp = "HOT"

                # ICE 선택
                elif t == 2:
                    self.temp = "ICE"

                # 잘못 입력
                else:
                    print("다시 입력하세요.")

            # 추가 주문 문자열 생성
            order = self.temp + ' ' + self.menu[n - 1]

            # 주문 리스트에 메뉴 추가
            self.order_menu.append(order)

            # 가격 리스트에 가격 추가
            self.order_price.append(self.price[n - 1])

            # 총 금액 추가
            self.price_sum += self.price[n - 1]

            # 추가 주문 출력
            print()

            print("추가 주문 음료",
                  order,
                  ':',
                  self.price[n - 1],
                  '원')

            print("합계 :",
                  self.price_sum,
                  '원')


        # 주문 완료 출력
        print("\n주문이 완료되었습니다.")

        # 주문표 출력 메소드 실행
        self.table()

        # 결제 메소드 실행
        self.pay()


    # 주문표 출력 메소드
    def table(self):

        # 줄 바꿈
        print("\n")

        # 주문표 상단 출력
        print("=" * 40)

        # 주문표 제목 가운데 정렬
        print("ORDER LIST".center(40))

        print("=" * 40)

        # 주문 리스트 개수만큼 반복
        for i in range(len(self.order_menu)):

            # 주문 번호 / 메뉴 / 가격 출력
            print(f"{i+1}. "
                  f"{self.order_menu[i]:<20}"
                  f"{self.order_price[i]:>6}원")

        # 구분선 출력
        print("-" * 40)

        # 총 합계 출력
        print(f"{'TOTAL':<25}"
              f"{self.price_sum:>7}원")

        # 주문표 하단 출력
        print("=" * 40)


    # 결제 메소드
    def pay(self):

        # 총 결제 금액 출력
        print(f"\n총 합계 금액은 {self.price_sum}원 입니다.")

        # 결제 방법 변수
        self.pay_method = 0

        # 1 또는 2를 입력할 때까지 반복
        while self.pay_method != 1 and self.pay_method != 2:

            # 결제 방법 입력
            self.pay_method = int(input(
                "현금 결제는 1번, 카드 결제는 2번을 입력해주세요. : "
            ))

            # 현금 결제
            if self.pay_method == 1:

                print("직원을 호출하겠습니다.")

            # 카드 결제
            elif self.pay_method == 2:

                print("IC칩 방향에 맞게 카드를 꽂아주세요.")

            # 잘못 입력
            else:

                print("다시 결제를 시도해 주세요.")


# 객체 생성
a = Kiosk()

# 메뉴 출력 메소드 실행
a.menu_print()

# 주문 메소드 실행
a.menu_select()
```


## 3. 프로젝트 확장하기

### 추가된 기능
- 메뉴와 가격 리스트를 다차원 리스트로 변경하기
- `enumerate()` 함수로 메뉴 출력하기
- `datetime` 모듈로 주문 일시 출력하기
- 데코레이터로 주문표 양식 출력하기


```python
# datetime 모듈 불러오기
from datetime import datetime


# 주문표 데코레이터 함수
def receipt_decorator(func):

    # wrapper 함수
    def wrapper(self):

        print("=" * 40)
        print("Maison de Café".center(40))
        print("ORDER LIST".center(40))
        print("=" * 40)

        # 원래 함수 실행
        func(self)

        print("=" * 40)

    return wrapper


# 키오스크 클래스 생성
class Kiosk:

    # 생성자 메소드
    def __init__(self):

        # 메뉴와 가격을 다차원 리스트로 저장
        self.menu = [
            ['Americano', 5800],
            ['Latte', 6800],
            ['Mocha', 7200],
            ['Yuzu_Tea', 7500],
            ['Green_Tea', 7300],
            ['Choco_Latte', 7800]
        ]

        # 주문 리스트
        self.order_menu = []

        # 총 금액
        self.price_sum = 0


    # 메뉴 출력 메소드
    def menu_print(self):

        print("=" * 40)
        print("Maison de Café".center(40))
        print("=" * 40)

        # enumerate() 함수 사용
        for i, item in enumerate(self.menu, start=1):

            print(f"{i}. "
                  f"{item[0]:<20}"
                  f"{item[1]:>6}원")

        print("=" * 40)


    # 주문 메소드
    def menu_select(self):

        # 주문 시간 저장
        self.order_time = datetime.now()

        # 메뉴 선택
        n = int(input("\n음료 번호를 입력하세요 : "))

        # 메뉴 번호 확인
        while n < 1 or n > len(self.menu):

            print("없는 메뉴입니다.")
            n = int(input("다시 입력하세요 : "))

        # HOT / ICE 선택
        t = 0

        while t != 1 and t != 2:

            t = int(input(
                "HOT 음료는 1을, ICE 음료는 2를 입력하세요 : "
            ))

            if t == 1:
                self.temp = "HOT"

            elif t == 2:
                self.temp = "ICE"

            else:
                print("다시 입력하세요.")

        # 주문 저장
        order = self.temp + ' ' + self.menu[n - 1][0]

        self.order_menu.append(
            [order, self.menu[n - 1][1]]
        )

        # 총 금액 추가
        self.price_sum += self.menu[n - 1][1]

        # 추가 주문
        while True:

            n = int(input(
                "\n추가 주문은 음료 번호를, 지불은 0을 누르세요 : "
            ))

            # 결제
            if n == 0:
                break

            # 메뉴 번호 확인
            if n < 1 or n > len(self.menu):

                print("없는 메뉴입니다.")
                continue

            # HOT / ICE 선택
            t = 0

            while t != 1 and t != 2:

                t = int(input(
                    "HOT 음료는 1을, ICE 음료는 2를 입력하세요 : "
                ))

                if t == 1:
                    self.temp = "HOT"

                elif t == 2:
                    self.temp = "ICE"

                else:
                    print("다시 입력하세요.")

            # 주문 저장
            order = self.temp + ' ' + self.menu[n - 1][0]

            self.order_menu.append(
                [order, self.menu[n - 1][1]]
            )

            # 총 금액 추가
            self.price_sum += self.menu[n - 1][1]


        print("\n주문이 완료되었습니다.")

        # 주문표 출력
        self.table()

        # 결제 메소드 실행
        self.pay()


    # 결제 메소드
    def pay(self):

        # 총 금액 출력
        print(f"\n총 합계 금액은 {self.price_sum}원 입니다.")

        # 결제 방법 변수
        self.pay_method = 0

        # 1 또는 2 입력 전까지 반복
        while self.pay_method != 1 and self.pay_method != 2:

            self.pay_method = int(input(
                "현금 결제는 1번, 카드 결제는 2번을 입력해주세요. : "
            ))

            # 현금 결제
            if self.pay_method == 1:

                print("직원을 호출하겠습니다.")

            # 카드 결제
            elif self.pay_method == 2:

                print("IC칩 방향에 맞게 카드를 꽂아주세요.")



    # 주문표 출력 메소드
    # 데코레이터 적용
    @receipt_decorator
    def table(self):

        # 주문 일시 출력
        print("주문 일시 :",
              self.order_time.strftime("%Y-%m-%d %H:%M:%S"))

        print("-" * 40)

        # 주문 내역 출력
        for i, item in enumerate(self.order_menu, start=1):

            print(f"{i}. "
                  f"{item[0]:<20}"
                  f"{item[1]:>6}원")

        print("-" * 40)

        # 총 금액 출력
        print(f"{'TOTAL':<25}"
              f"{self.price_sum:>7}원")


# 객체 생성
a = Kiosk()

# 메뉴 출력
a.menu_print()

# 주문 실행
a.menu_select()
```
